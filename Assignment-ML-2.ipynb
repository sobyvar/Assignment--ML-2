```python
# Problem:-1 Build the linear regression model using scikit learn in boston data to predict 'Price' based on other dependent variable.
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import load_boston
boston = load_boston()
bos = pd.DataFrame(boston.data)
bos.head()
```

```python
# Renaming the columns with proper names
bos.columns = boston.feature_names
bos.head()
```

```python
# Adding the price to the dataframe
bos['PRICE']=boston.target
bos.head()
```

```python
# Print the intercept and the coefficient
from sklearn.linear_model import LinearRegression
X= bos.drop('PRICE',axis =1)

lm =LinearRegression()
lm.fit(X,bos.PRICE)
print('1.Estimated intercept:',lm.intercept_)
print('2.Estimated Coefficients are :',lm.coef_)
```

```python
# Print all the columns with their coefficient
pd.DataFrame(list(zip(X.columns,lm.coef_)), columns = ['features', 'estimatedCoefficients'])
```

```python
print('The predicted price is')
lm.predict(X)
```

```python
# Plot the Price vs Predicted Price
plt.scatter(bos.PRICE,lm.predict(X))
plt.xlabel("Prices:'$Y_i$")
plt.ylabel("Model Predicted Prices:'$Y\hat{Y}_i$")
plt.title("Prices vs Model Predicted Prices: $Y_i$ vs $Y\hat{Y}_i$")# Calculate the means square error
mse = np.mean((bos.PRICE - lm.predict(X))**2)
mse
```
